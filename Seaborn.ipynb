```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")
print(tips.head())
```

```python
import matplotlib.pyplot as plt
sns.scatterplot(data=tips, x="total_bill", y="tip")
plt.show()
```

```python
sns.lineplot(data=tips, x="total_bill", y="tip")
plt.show()
```

```python
sns.barplot(data=tips, x="day", y="tip", estimator='mean')
plt.show()
```

```python
sns.countplot(data=tips, x="day")
plt.show()

```

```python
sns.boxplot(data=tips, x="day", y="total_bill")
plt.show()
```

```python
sns.violinplot(data=tips, x="day", y="total_bill")
plt.show()
```

```python
sns.histplot(data=tips, x="total_bill", kde=True)
plt.show()
```

```python
iris = sns.load_dataset("iris")
sns.pairplot(iris)
plt.show()
```

```python
titanic = sns.load_dataset("titanic")
sns.heatmap(titanic.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.show()
```

```python
penguins = sns.load_dataset("penguins")
sns.scatterplot(data=penguins, x="bill_length_mm", y="flipper_length_mm")
plt.show()
```

```python
sns.boxplot(data=titanic, x="class", y="fare")
plt.show()
```

```python
sns.countplot(data=titanic, x="survived")
plt.show()
```

```python
g = sns.FacetGrid(tips, col="sex")
g.map(plt.hist, "total_bill")
plt.show()
```

```python
sns.heatmap(penguins.corr(numeric_only=True), annot=True, cmap='YlGnBu')
plt.show()
```

```python
diamonds = sns.load_dataset("diamonds")
sns.violinplot(data=diamonds, x="cut", y="price")
plt.show()
```

```python
sns.pairplot(penguins, hue="species")
plt.show()
```

```python
flights = sns.load_dataset("flights")
flights_pivot = flights.pivot(index="month", columns="year", values="passengers")
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu")
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")
tips_numeric = tips.copy()
for col in tips_numeric.columns:
    if tips_numeric[col].dtype == 'object' or pd.api.types.is_categorical_dtype(tips_numeric[col]):
        tips_numeric[col], _ = pd.factorize(tips_numeric[col])
correlation = tips_numeric.corr(numeric_only=True)
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.show()
```
